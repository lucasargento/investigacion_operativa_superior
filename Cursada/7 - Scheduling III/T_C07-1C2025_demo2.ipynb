```python
from ortools.sat.python.cp_model import CpModel, CpSolver

m = CpModel()

max_ub =1000000

# CARGA DE DATOS

nec = {
    'VEZMU1EQ': 134,
    'QTROUV79': 123,
    'YUIIIP4W': 220,
}

prod = {
    'VEZMU1EQ': 
    {
    'corte_m': 48,
    'doblado': 38,
    'soldado': 56,
    'corte_w': 0,
    'tapizado': 0,
    'pintura': 36,
    'despacho': 20
    },
    'QTROUV79': 
    {
    'corte_m': 45,
    'doblado': 23,
    'soldado': 32,
    'corte_w': 34,
    'tapizado': 102,
    'pintura': 36,
    'despacho': 34
    },
    'YUIIIP4W': 
    {
    'corte_m': 0,
    'doblado': 0,
    'soldado': 0,
    'corte_w': 23,
    'tapizado': 0,
    'pintura': 15,
    'despacho': 24
    }
}

op = ['corte_m','doblado','soldado','corte_w','tapizado','pintura','despacho']

ultima_op = 'despacho'

ope = [] # lista que contiene todos los intervalos

# definicion de intervalos

for i in nec:
    r = []  # lista que contendr치 los intervalos para un producto determinado
    for j in op:     
        r.append(
            m.new_interval_var(
                start=m.NewIntVar(lb=0, ub=max_ub, name='start_'+j + '_' + i), 
                end=m.NewIntVar(lb=0, ub=max_ub, name='end_'+j + '_' + i), 
                size = prod[i][j]*nec[i], 
                name = j + '_' + i))
    ope.append(r)

# definici칩n de relaciones de precedencia

for i in range(len(nec)):
    for j in range(len(op)-1):
        m.add(ope[i][j].end_expr() <= ope[i][j+1].start_expr())

# no overlap

for j in range(len(op)):
    m.add_no_overlap([ ope[i][j] for i in range(len(nec))])

# funcional

obj = m.new_int_var(0, max_ub, 'obj')
m.add_max_equality(obj, [ope[i][op.index(ultima_op)].end_expr() for i in range(len(nec)) ])
m.minimize(obj)

# resoluci칩n

s = CpSolver()
e = s.solve(m)

# resultados

if e == 4 or e == 2:
    print("estado: ", e) # 4 es optimal
    print("objetivo: ", s.objective_value) 
    print("\n")
    for i in range(len(nec)):
        for j in op:
            print("producto:", i, "operacion:", j, 'inicio:', s.value(ope[i][op.index(j)].start_expr()), 'fin:', s.value(ope[i][op.index(j)].end_expr()) )
        print("\n")
```
