### **DEMO 1**: Trabajo con imágenes

```python
import cv2                                              # OpenCV (herramientas de visión por computadora)
import numpy as np  
import matplotlib.pyplot as plt
```

```python
casa = cv2.imread('imgs/T_C09-1C2025_demo1_casa.jpg')   # casa es una matriz de numpy con los valores de cada pixel
```

```python
casa.shape                                              # vemos la forma de la matriz
```

```python
plt.imshow(casa) 
plt.show()
```

```python
# # OpenCV trabaja con BGR pero matplotlib en RGB, por lo anterior modificamos el mapa de color (la ubicacion de las matrices)
plt.imshow(  cv2.cvtColor(casa, cv2.COLOR_BGR2RGB) ) 
plt.show()
```

```python
casa_bn = cv2.cvtColor(casa, cv2.COLOR_BGR2GRAY)    # cambiamos la imagen a escala de grises
print(casa_bn.shape)

plt.imshow(cv2.cvtColor(casa_bn, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
x = 400
y = 380
wx = 35
wy = 35

recorte = casa_bn[ y : y + wy, x : x + wx ]                         # genero el recorte de 35x35

casa_b = cv2.rectangle(casa, (x,y),(x+wx,y+wy), (0,0,255), 3)       # dibujo un rectagulo con el recorrte en una nueva imagen

plt.imshow(cv2.cvtColor(casa_b, cv2.COLOR_BGR2RGB))
plt.show()
```
