```python
import cv2
import matplotlib.pyplot as plt
```

#### **BORRADOR PARA COORDENADAS DEL INDICADOR LUMINOSO**

```python
cap = cv2.VideoCapture(1)   # verificar el puerto de la c√°mara
```

```python

_, img = cap.read()                 # leo una imagen

img_c = img.copy()                  # copio la imagen ... prestar atenci√≥n que no es tan simple como igualar

x = 1000
y = 525

frame = cv2.line(img_c, (0,y),(img.shape[1], y), (0,0,255), 5)      # linea horizontal
frame = cv2.line(img_c, (x,0),(x, img.shape[0]), (0,0,255), 5)      # linea vertical

# muestro la imagen
plt.imshow(cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
# imprimo los valores del punto encontrado
print(img[y,x])
```
