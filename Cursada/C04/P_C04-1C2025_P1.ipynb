# PASO 1: MODELO BÁSICO

```python
# MODELO

from docplex.cp.model import CpoModel

m = CpoModel()

# L0421
L0421_Y4W_CM = m.interval_var(name = "L0421_Y4W_CM", size = 220*23)
L0421_Y4W_PI = m.interval_var(name = "L0421_Y4W_PI", size = 220*15)
L0421_Y4W_DE = m.interval_var(name = "L0421_Y4W_DE", size = 220*24)

# L0422
L0422_Q79_MC = m.interval_var(name = "L0422_Q79_MC", size = 123*45) 
L0422_Q79_DM = m.interval_var(name = "L0422_Q79_DM", size = 123*23)
L0422_Q79_SO = m.interval_var(name = "L0422_Q79_SO", size = 123*32)
L0422_Q79_CM = m.interval_var(name = "L0422_Q79_CM", size = 123*34)
L0422_Q79_TA = m.interval_var(name = "L0422_Q79_TA", size = 123*102)
L0422_Q79_PI = m.interval_var(name = "L0422_Q79_PI", size = 123*36)
L0422_Q79_DE = m.interval_var(name = "L0422_Q79_DE", size = 123*34)

# L0423
L0423_VEQ_MC = m.interval_var(name = "L0423_VEQ_MC", size = 134*48)
L0423_VEQ_DM = m.interval_var(name = "L0423_VEQ_DM", size = 134*38)
L0423_VEQ_SO = m.interval_var(name = "L0423_VEQ_SO", size = 134*56)
L0423_VEQ_PI = m.interval_var(name = "L0423_VEQ_PI", size = 134*36)
L0423_VEQ_DE = m.interval_var(name = "L0423_VEQ_DE", size = 134*20)

# FUNCIONAL

m.add(m.minimize(m.end_of(L0421_Y4W_DE)+m.end_of(L0422_Q79_DE)+m.end_of(L0423_VEQ_DE)))   

# RELACIONES DE PRECEDENCIA

# L0421
m.add(m.end_of(L0421_Y4W_CM) <= m.start_of(L0421_Y4W_PI))
m.add(m.end_of(L0421_Y4W_PI) + 24*60 <= m.start_of(L0421_Y4W_DE))

# L0422
m.add(m.end_of(L0422_Q79_MC) <= m.start_of(L0422_Q79_DM))
m.add(m.end_of(L0422_Q79_DM) <= m.start_of(L0422_Q79_SO))
m.add(m.end_of(L0422_Q79_SO) == m.start_of(L0422_Q79_PI))
m.add(m.end_of(L0422_Q79_CM) <= m.start_of(L0422_Q79_TA))
m.add(m.end_of(L0422_Q79_TA) <= m.start_of(L0422_Q79_PI))
m.add(m.end_of(L0422_Q79_PI) + 24*60 <= m.start_of(L0422_Q79_DE))

# L0423
m.add(m.end_of(L0423_VEQ_MC) <= m.start_of(L0423_VEQ_DM))
m.add(m.end_of(L0423_VEQ_DM) <= m.start_of(L0423_VEQ_SO))
m.add(m.end_of(L0423_VEQ_SO) == m.start_of(L0423_VEQ_PI))
m.add(m.end_of(L0423_VEQ_PI) + 24*60 <= m.start_of(L0423_VEQ_DE))

s = m.solve()

if s:
    s.print_solution()
else:
    print('no se encuentra solución.')

```
