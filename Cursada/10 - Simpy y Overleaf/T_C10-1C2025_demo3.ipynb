### **DEMO 3**: CONTENEDORES

```python
import simpy as sim
import numpy as np
```

```python
# source -> queue (cap 10) -> delay (cap 1) -> sink

def proceso(env, c, fila):
    r = operario.request()
    yield r
    yield fila.get(1)                                               # NUEVO
    print(f'-> entra item {c} en {env.now:.2f}')
    yield env.timeout(np.random.triangular(1,4,6))
    print(f'<- sale item {c} en {env.now:.2f}')
    operario.release(r)
    
def principal(env, fila):
    c = 0
    while True:    
        yield env.timeout(np.random.exponential(1.7))
        print(f'[tamaño fila {fila.level}]')                        # NUEVO
        if fila.level<fila.capacity:                                # NUEVO
            fila.put(1)                                             # NUEVO
            print(f'*->: entrada al sistema en {env.now:.2f}')      # NUEVO
            c += 1
            env.process(proceso(env, c, fila))                      # NUEVO
        else:
            print('[Fila llena]')                                   # NUEVO

env = sim.Environment()
operario = sim.Resource(env=env, capacity=1)
fila = sim.Container(env=env, init=0, capacity=10)                  # NUEVO
env.process(principal(env, fila))
env.run(until=100)
```
