#### inicialmente, de la clase anterior ...

```python
# source -> delay -> sink

def proceso(env):
    yield env.timeout(np.random.triangular(1,4,6))

def principal(env):
    while True:
        yield env.timeout(np.random.exponential(1.7))
        env.process(proceso(env))

env = sim.Environment()
env.process(principal(env))
env.run(until=100)
```

```python
# versiÃ³n con impresiones ...

def proceso(env, c):
    print(f'entra {c}')
    yield env.timeout(np.random.triangular(1,4,6))
    print(f'sale {c}')
  
def principal(env):
    c = 0
    while True:
        yield env.timeout(np.random.exponential(1.7))
        c += 1
        env.process(proceso(env, c))

env = sim.Environment()
env.process(principal(env))
env.run(until=100)
```

#### **DEMO 2**: USO DE RECURSOS

```python
#source -> delay (cap 1) -> sink

def proceso(env, c):
    req = operadores.request()                                     # NUEVO
    yield req                                                      # NUEVO
    print(f'-> entra item {c} en {env.now:.2f}')
    yield env.timeout(np.random.triangular(1,4,6))
    print(f'<- sale item {c} en {env.now:.2f}')
    operadores.release(req)                                        # NUEVO
  
def principal(env):
    c = 0
    while True:    
        yield env.timeout(np.random.exponential(1.7))
        c += 1
        env.process(proceso(env, c))

env = sim.Environment()
operadores = sim.Resource(env, capacity=1)                         # NUEVO
env.process(principal(env))
env.run(until=1000)
```

#### **DEMO 2**: USO DE RECURSOS C/ IMPACIENCIA

```python
def proceso(env, c):
    req = maquina.request()
    yield req | env.timeout(20)                                     # NUEVO (CAMBIO)
    if req.triggered:                                               # NUEVO
        print(f'-> entra item {c} en {env.now:.2f}')
        yield env.timeout(np.random.triangular(1,4,6))
        print(f'<- sale item {c} en {env.now:.2f}')
        maquina.release(req)                                        
    else:                                                           # NUEVO
        print(f'** timeout item {c} en {env.now:.2f}')              # NUEVO      
        req.cancel()                                                # NUEVO
        
  
def principal(env):
    c = 0
    while True:    
        yield env.timeout(np.random.exponential(1.7))
        c += 1
        print(f'-> llega item {c} en {env.now:.2f}')              # NUEVO
        env.process(proceso(env, c))

env = sim.Environment()
maquina = sim.Resource(env, capacity=1)                         
env.process(principal(env))
env.run(until=1000)
```
